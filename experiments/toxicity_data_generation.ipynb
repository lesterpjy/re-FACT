## Combine names with toxic sentences

```python
# -----------------------------------------------------------------------------
# 2. Sort df by toxicity
# -----------------------------------------------------------------------------
df_sorted = df.sort_values(by='toxicity', ascending=True).reset_index(drop=True)
# Now df_sorted['toxicity'] is in ascending order

# -----------------------------------------------------------------------------
# 3. Group names_df by country
#    We’ll create a structure that tells us how many rows each country needs
# -----------------------------------------------------------------------------
country_groups = names_df.groupby('country')
countries = country_groups.groups.keys()

# We'll keep track of:
# - how many texts each country needs (the number of names in that country)
# - an assignment of df indices (rows) to those countries
country_text_assignments = {}

# Create a list of all row indices in df_sorted (0 .. len(df_sorted)-1)
# so we can "pull" them as we assign to countries
available_text_indices = list(range(len(df_sorted)))

# -----------------------------------------------------------------------------
# 4. Balancing each country's average toxicity around 0.5
# -----------------------------------------------------------------------------
# We'll do a simple "greedy" two-pointer approach:
#   - pointer_low starts at the beginning of df_sorted (lowest toxicity)
#   - pointer_high starts at the end of df_sorted (highest toxicity)
#   - we try to pick from the low end if we are above target, or from the high end if below target,
#     to nudge the average toxicity toward 0.5
#
#   This is not guaranteed to be perfect, but should produce a somewhat balanced distribution.
#
pointer_low = 0
pointer_high = len(df_sorted) - 1

country_text_assignments = {country: [] for country in countries}

for country in countries:
    # Number of texts needed for this country
    n_needed = len(country_groups.get_group(country))
    
    # Start with an empty assignment
    assignments = []
    
    # We'll do a quick approach: fill from both ends, measuring partial average
    current_toxicities = []
    
    for _ in range(n_needed):
        if len(current_toxicities) == 0:
            # If we haven't assigned anything yet, just pick from the middle-ish
            mid_idx = (pointer_low + pointer_high) // 2
            assignments.append(mid_idx)
            current_toxicities.append(df_sorted.loc[mid_idx, 'toxicity'])
            # Remove that index from the "pool" so we don't reuse it
            # We'll do so by marking pointer_low or pointer_high if it matches mid_idx
            # But simpler is to keep a set or list of “used” indices. For brevity, we just
            # increment pointer_low (knowing mid_idx ~ pointer_low or pointer_high).
            pointer_low += 1
        else:
            avg_so_far = np.mean(current_toxicities)
            if avg_so_far > 0.5:
                # if we're above 0.5, pick from the low-toxicity end to try to bring it down
                assignments.append(pointer_low)
                current_toxicities.append(df_sorted.loc[pointer_low, 'toxicity'])
                pointer_low += 1
            else:
                # if we're below 0.5, pick from the high-toxicity end
                assignments.append(pointer_high)
                current_toxicities.append(df_sorted.loc[pointer_high, 'toxicity'])
                pointer_high -= 1
    
    country_text_assignments[country] = assignments

# -----------------------------------------------------------------------------
# 5. Now we have an index assignment for each country, but we need to line them
#    up with the names in names_df in the same order the names appear.
# -----------------------------------------------------------------------------
# We'll build a helper so that for each row in names_df, we pop one assigned index
# from the relevant country’s list. (pop from the front of the list).
# -----------------------------------------------------------------------------
def get_next_text_for_country(ctry):
    return country_text_assignments[ctry].pop(0)  # pop from front

# Create new columns in names_df
names_df['assigned_text'] = None
names_df['assigned_toxicity'] = None
names_df['new_sentence'] = None

# We'll iterate over names_df in its current order
for i, row in names_df.iterrows():
    ctry = row['country']
    # get an index from df_sorted assigned to this country
    df_idx = get_next_text_for_country(ctry)
    
    chosen_text = df_sorted.loc[df_idx, 'text']
    chosen_tox = df_sorted.loc[df_idx, 'toxicity']
    
    names_df.at[i, 'assigned_text'] = chosen_text
    names_df.at[i, 'assigned_toxicity'] = chosen_tox
    names_df.at[i, 'new_sentence'] = f"{row['name']} wrote in the chat, '{chosen_text}'"

# -----------------------------------------------------------------------------
# 6. Check results
# -----------------------------------------------------------------------------
names_df.sort_values("assigned_toxicity")
```

```python
# confirm toxicity range is balanced
print("\nAverage toxicity by country:")
names_df.groupby('country')['assigned_toxicity'].mean()
```
